# Test VietnamMarketDataService Data Query

Query FPT daily bars from `2014-01-01` inclusive to `2015-01-01` exclusive through the service API.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd()
if not (PROJECT_ROOT / "src").exists():
    PROJECT_ROOT = PROJECT_ROOT.parent

sys.path.insert(0, str(PROJECT_ROOT / "src"))
PROJECT_ROOT
```

```python
import json
import os
from urllib.request import urlopen

from dotenv import load_dotenv

load_dotenv(PROJECT_ROOT / ".env")

BASE_URL = os.getenv("VN_MARKET_DATA_API_URL", "http://127.0.0.1:8000").rstrip("/")
health = json.loads(urlopen(f"{BASE_URL}/api/v1/health", timeout=10).read().decode("utf-8"))
health
```

```python
from loaders.market_data_api import load_market_bars

df = load_market_bars(
    symbols="FPT",
    start="2026-06-01",
    end="2026-06-19",
)

df.head()
```

```python
df.tail()
```

```python
summary = {
    "rows": len(df),
    "symbols": sorted(df["symbol"].dropna().unique().tolist()) if not df.empty else [],
    "min_time": df["time"].min() if not df.empty else None,
    "max_time": df["time"].max() if not df.empty else None,
}
summary
```

```python
if df.empty:
    raise AssertionError("No FPT rows returned. Check that the service is running and MongoDB contains FPT data for 2026.")

assert set(df["symbol"]) == {"FPT"}
assert df["time"].min() >= "2026-06-01"
assert df["time"].max() < "2026-06-19"
assert df["time"].is_monotonic_increasing

"FPT 2026 data query passed"
```
